```python
import spectral as spy
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize
```

```python
# === Load Hyperspectral Image ===
img = spy.open_image("cuprite.95.cal.rtgc.v.hdr")  # Replace with your .hdr file
data = img.load()
rows, cols, bands = data.shape
reshaped = data.reshape((rows * cols, bands))
```

```python
# === Normalize Spectra ===
reshaped = np.nan_to_num(reshaped)
reshaped = normalize(reshaped, axis=1)
# === Dimensionality Reduction (for clustering only) ===
pca = PCA(n_components=3)
reduced = pca.fit_transform(reshaped)
```

```python
# Function to classify minerals using SpectralGPT
def classify_minerals(spectral_image):
    height, width, _ = spectral_image.shape
    mineral_maps = {}
    minerals = ['Zn', 'Au', 'As', 'Ni', 'Cu', 'Hg', 'Fe2O3', 'TiO2', 'MnO']  # Example minerals

    for mineral in minerals:
        inputs = tokenizer(f"Detect {mineral} in image", return_tensors="pt")
        with torch.no_grad():
            outputs = model(**inputs)
        scores = outputs.logits.squeeze().numpy()
        mineral_map = (scores.reshape(height, width) > 0.5).astype(np.uint8) * 255  # Thresholding
        mineral_maps[mineral] = mineral_map

    return mineral_maps
```

```python
# === KMeans Clustering ===
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
labels = kmeans.fit_predict(reduced)
label_map = labels.reshape(rows, cols)

# === Extract Mean Spectrum for Each Cluster ===
mean_spectra = np.zeros((n_clusters, bands))
for i in range(n_clusters):
    mean_spectra[i] = reshaped[labels == i].mean(axis=0)

# === Sample Mineral Reference Spectra (normalized) ===
# Replace with real spectral libraries if available
# Each row = simulated spectral signature of a known mineral
mineral_refs = {
    'Kaolinite': np.linspace(0.3, 0.6, bands),
    'Alunite': np.linspace(0.2, 0.7, bands),
    'Hematite': np.linspace(0.1, 0.9, bands),
    'Chlorite': np.linspace(0.4, 0.5, bands),
    'Montmorillonite': np.linspace(0.5, 0.6, bands),
}
for key in mineral_refs:
    mineral_refs[key] = normalize(mineral_refs[key].reshape(1, -1))[0]
```

```python
# === Spectral Angle Mapper (SAM) ===
def spectral_angle(vec1, vec2):
    dot = np.dot(vec1, vec2)
    norms = np.linalg.norm(vec1) * np.linalg.norm(vec2)
    return np.arccos(np.clip(dot / norms, -1.0, 1.0))  # result in radians

# Match each cluster to closest mineral
cluster_mineral_names = []
for spectrum in mean_spectra:
    best_match = None
    smallest_angle = float('inf')
    for mineral, ref_spectrum in mineral_refs.items():
        angle = spectral_angle(spectrum, ref_spectrum)
        if angle < smallest_angle:
            smallest_angle = angle
            best_match = mineral
    cluster_mineral_names.append(best_match)
```

```python
# === Create Mineral Map ===
mineral_map = np.vectorize(lambda x: cluster_mineral_names[x])(label_map)
```

```python
# Map to colors
unique_minerals = list(set(cluster_mineral_names))
mineral_colors = {m: i for i, m in enumerate(unique_minerals)}
colored_map = np.vectorize(lambda x: mineral_colors[x])(mineral_map)
```

```python
# === Plot ===
plt.figure(figsize=(10, 6))
plt.imshow(colored_map, cmap='tab10')
plt.title("Mineral Classification Map (SAM + KMeans)")
cbar = plt.colorbar(ticks=range(len(unique_minerals)))
cbar.ax.set_yticklabels(unique_minerals)
plt.axis('off')
plt.show()
```

```python
def mock_spectral_gpt_encoder(spectrum):
    # In practice, replace this with your model's embedding
    spectrum = normalize(spectrum.reshape(1, -1))[0]
    return np.fft.fft(spectrum).real[:128]  # Mock embedding (128D)
```

```python
# === KMeans Clustering
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
labels = kmeans.fit_predict(reduced)
label_map = labels.reshape((rows, cols))
```

```python
# === Extract Cluster Mean Spectra
mean_spectra = np.zeros((n_clusters, bands))
for i in range(n_clusters):
    mean_spectra[i] = reshaped[labels == i].mean(axis=0)
```

```python
# === Get Embeddings for Cluster Spectra
cluster_embeddings = [mock_spectral_gpt_encoder(spectrum) for spectrum in mean_spectra]

# === Define Reference Mineral Spectra (Mock Spectra for Demo)
# Replace these with real mineral spectra from USGS or a spectral library
mineral_refs = {
    'Kaolinite': np.linspace(0.3, 0.6, bands),
    'Alunite': np.linspace(0.2, 0.7, bands),
    'Hematite': np.linspace(0.1, 0.9, bands),
    'Chlorite': np.linspace(0.4, 0.5, bands),
    'Montmorillonite': np.linspace(0.5, 0.6, bands),
}
```

```python
# === Get Embeddings for Reference Minerals
mineral_embeddings = {
    name: mock_spectral_gpt_encoder(spec)
    for name, spec in mineral_refs.items()
}
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# === Match Clusters to Minerals using Cosine Similarity
mineral_labels = []
for cluster_emb in cluster_embeddings:
    best_match = max(
        mineral_embeddings.items(),
        key=lambda item: cosine_similarity([cluster_emb], [item[1]])[0][0]
    )
    mineral_labels.append(best_match[0])
```

```python
# === Create Classified Mineral Map
mineral_map = np.vectorize(lambda x: mineral_labels[x])(label_map)
unique_minerals = list(set(mineral_labels))
mineral_color_map = {m: i for i, m in enumerate(unique_minerals)}
colored_map = np.vectorize(lambda x: mineral_color_map[x])(mineral_map)
```

```python
# === Plot Result
plt.figure(figsize=(10, 6))
plt.imshow(colored_map, cmap='tab10')
plt.title("Mineral Classification Map (Spectral GPT + KMeans)")
cbar = plt.colorbar(ticks=range(len(unique_minerals)))
cbar.ax.set_yticklabels(unique_minerals)
plt.axis('off')
plt.show()
```

```python
gpt_image_path = "Spectral_Kmeans.jpg"
sam_image_path = "SAM_Kmeans.jpg"
```

```python
from PIL import Image
gpt_img = np.array(Image.open(gpt_image_path))
sam_img = np.array(Image.open(sam_image_path))
```

```python
# Function to convert RGB image to cluster labels
def image_to_labels(img, n_clusters=10):
    if img.ndim == 3:
        pixels = img.reshape(-1, 3)
        kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(pixels)
        labels = kmeans.labels_.reshape(img.shape[:2])
    else:
        labels = img  # Assume already labeled
    return labels
```

```python
# Function to find the dominant cluster
def get_dominant_cluster(labels):
    values, counts = np.unique(labels, return_counts=True)
    return values[np.argmax(counts)]

# Function to highlight the dominant cluster
def highlight_cluster(labels, dominant_cluster):
    mask = (labels == dominant_cluster).astype(np.uint8) * 255
    return mask
```

```python
# Process both images
labels_gpt = image_to_labels(gpt_img)
labels_sam = image_to_labels(sam_img)

dom_gpt = get_dominant_cluster(labels_gpt)
dom_sam = get_dominant_cluster(labels_sam)

highlight_gpt = highlight_cluster(labels_gpt, dom_gpt)
highlight_sam = highlight_cluster(labels_sam, dom_sam)
```

```python
# Plot the highlighted dominant clusters
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(highlight_gpt, cmap="Reds")
axes[0].set_title("Spectral GPT Dominant Cluster (Alunite?)")
axes[0].axis('off')

axes[1].imshow(highlight_sam, cmap="Blues")
axes[1].set_title("SAM Dominant Cluster (Chlorite?)")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
